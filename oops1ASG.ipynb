```python
import logging as log
log.basicConfig(filename=r"C:\Users\ragha\Desktop\coding\pw\assignments\oopsASG.log", level=log.DEBUG, format="%(asctime)s - %(levelname)s - %(message)s")
# Problem 1: Bank Account Create a class representing a bank account with attributes like account number, account holder name, and balance. Implement methods to deposit and withdraw money from the account.
class Bankaccount() :
    def __init__(self, accNum, holderName, balance) :
        self.accountNumber = accNum
        self.nameOfHolder = holderName
        self.balance = balance
    
    def deposit(self, amt) :
        try :
            log.info("User is depositing cash")
            self.balance += float(amt)
            return f"{float(amt)} rupees deposited in {self.accountNumber}. Final Balance -> {self.balance}"
        except (ValueError, TypeError):
            log.error("User did not enter a number")
            return "Enter a number"
    
    def withdraw(self, amt) :
        try :
            log.info("User is withdrawing cash")
            if amt > self.balance :
                raise Exception()
            self.balance -= amt
            return f"{float(amt)} rupees withdrawn from {self.accountNumber}. Final Balance -> {self.balance}"
        except (ValueError, TypeError):
            log.error("User did not enter a number")
            return "Enter a number"
        except Exception :
            log.error("User entered more amount to withdraw than he had")
            return "You don't have that amount of money in your account"

acc1 = Bankaccount(1001001001, "Raghav", 1000000)
print(acc1.deposit(100000))
print(acc1.deposit("a"))
print(acc1.withdraw(500000))
print(acc1.withdraw("a"))
print(acc1.withdraw(1500000))


# Problem 2: Employee Management Create a class representing an employee with attributes like employee ID, name, and salary. Implement methods to calculate the yearly bonus and display employee details.
class Employee() :
    def __init__(self, id : int, name : str, salary : int) :
        self.id = id
        self.name = name
        self.salary = salary
    
    def display_details(self) :
        log.info("Displaying Employee Details")
        return f"Name : {self.name}, ID : {self.id}, Salary : {self.salary}"
    
    def bonus(self) :
        log.info("Calculating bonus")
        self.salary += (0.05 * self.salary)
        return f"Salary with bonus {self.salary}"
user1 = Employee(2129, "Deepak", 500000)
print(user1.display_details())
print(user1.bonus())

# Problem 3: Vehicle Rental Create a class representing a vehicle rental system. Implement methods to rent a vehicle, return a vehicle, and display available vehicles.
class vehicleRental() :
    available = [    
        "Ford Mustang",
        "Chevrolet Corvette",
        "Porsche 911"
    ]

    rented = []
    def display(self) :
        log.info("Showing the user list of available ")
        return f"Available  : {self.available}"
    
    def rentCar(self, carToRent, time_duration_hrs) :
        try :
            log.info("Trying to rent a car for the user")
            if len(self.available) == 0:
                raise IndexError()
            i = self.available.index(carToRent)
            self.available.pop(i)
        except ValueError :
            log.error("Car not available")
            return f"Sorry! {carToRent} is not available at the moment"
        except IndexError :
            log.error("No car available")
            return "Sorry! No  available at the moment"
        else :
            log.info("Car successfully rented")
            self.rented.append(carToRent)
            return f"Thank you for renting the car! Return it within {time_duration_hrs} hours. You will be billed accordingly"
    
    def returnCar(self, carToReturn) :
        try :
            log.info("Trying to return the car")
            self.rented.remove(carToReturn)
            self.available.append(carToReturn)
        except ValueError :
            log.error("Not our car")
            return "This is not our car"
        else :
            log.info("Car returned successfully")
            return "Thank you! Car returned successfully"
customer1 = vehicleRental()
print(customer1.display())
print(customer1.rentCar("Ford Mustang", 5))
customer2 = vehicleRental()
print(customer2.rentCar("Ford Mustang", 4))
print(customer2.rentCar("Porsche 911", 4))
customer3 = vehicleRental()
print(customer3.rentCar("Chevrolet Corvette", 5))
customer4 = vehicleRental()
print(customer4.rentCar("ABC", 7))
print(customer2.returnCar("Porsche 911"))
print(customer1.returnCar("Ford Mustang"))
print(customer3.returnCar("Chevrolet Corvette"))
print(customer4.display())


# Problem 4: Library Catalog Create classes representing a library and a book. Implement methods to add books to the library, borrow books, and display available books.
class library() :
    availablebooks = [    
        "Python for dummies",
        "DSA for dummies",
        "Harry Potter Series"
    ]

    rentedbooks = []
    def display(self) :
        log.info("Showing the user list of available books")
        return f"Available books : {self.availablebooks}"
    
    def rentbook(self, bookToRent, time_duration_weeks) :
        try :
            log.info("Trying to rent a book for the user")
            if len(self.availablebooks) == 0:
                raise IndexError()
            i = self.availablebooks.index(bookToRent)
            self.availablebooks.pop(i)
        except ValueError :
            log.error("book not available")
            return f"Sorry! {bookToRent} is not available at the moment"
        except IndexError :
            log.error("No book available")
            return "Sorry! No books available at the moment"
        else :
            log.info("book successfully rented")
            self.rentedbooks.append(bookToRent)
            return f"Thank you for renting the book! Return it within {time_duration_weeks} weeks. You will be billed accordingly"
    
    def returnbook(self, bookToReturn) :
        try :
            log.info("Trying to return the book")
            self.rentedbooks.remove(bookToReturn)
            self.availablebooks.append(bookToReturn)
        except ValueError :
            log.error("Not our book")
            return "This is not our book"
        else :
            log.info("book returned successfully")
            return "Thank you! book returned successfully"
customer1 = library()
print(customer1.display())
print(customer1.rentbook("Python for dummies", 2))
customer2 = library()
print(customer2.rentbook("Python for dummies", 1))
print(customer2.rentbook("DSA for dummies", 1))
customer3 = library()
print(customer3.rentbook("Harry Potter Series", 3))
customer4 = library()
print(customer4.rentbook("ABC", 1))
print(customer2.returnbook("DSA for dummies"))
print(customer1.returnbook("Python for dummies"))
print(customer3.returnbook("Harry Potter Series"))
print(customer4.display())

# Problem 5: Product Inventory Create classes representing a product and an inventory system. Implement methods to add products to the inventory, update product quantity, and display available products.
class Inventory() :
    availableproducts = [    
        "A1",
        "A2",
        "B1",
        "B2",
        "C1",
    ]

    def display(self) :
        log.info("Showing the list of available products")
        return f"Available products : {self.availableproducts}, Number of products : {len(self.availableproducts)}"
    
    def addProduct(self, productToAdd) :
        try :
            log.info("Trying to add a product")
            self.availableproducts.append(productToAdd)
        except Exception :
            log.error("Something went wrong")
            return f"Something went wrong"
        else :
            log.info("product successfully added")
            return "product successfully added"
    
checker1 = Inventory()
print(checker1.display())
print(checker1.addProduct("C2"))
print(checker1.display())

# Problem 6: Shape Calculation Create a class representing a shape with attributes like length, width, and height. Implement methods to calculate the area and perimeter of the shape.
class Shape() :
    def __init__(self, length, width, height) :
        self.length = length
        self.width = width
        self.height = height
    
    def area(self) :
        # Area is for 2d shapes and volume is used for 3d shapes
        try :
            log.info("Calculating Area")
            if self.height < 0 or self.width <= 0 or self.length <= 0 :
                raise Exception("Values can't be negative and length/width can't be 0")
            if self.height <= 1 : #basically a 2d shape
                area = self.length * self.width
            area = self.length * self.width * self.height
        except Exception :
            log.error("Values can't be negative and length/width can't be 0")
            return "Values can't be negative and length/width can't be 0"
        else :
            return f"Area of the shape is {area}"
    
    def perimeter(self) :
        try :
            log.info("Calculating Perimeter")
            if self.height < 0 or self.width <= 0 or self.length <= 0 :
                raise Exception("Values can't be negative and length/width can't be 0")
            if self.height <= 1 : #basically a 2d shape
                perimeter = 2 * (self.length + self.width)
            perimeter = 2 * ((self.length * self.width) + (self.length * self.height) + (self.height * self.width))
        except Exception :
            log.error("Values can't be negative and length/width can't be 0")
            return "Values can't be negative and length/width can't be 0"
        else :
            return f"perimeter of the shape is {perimeter}"

shape1 = Shape(3, 4, 0)
print(shape1.area())
print(shape1.perimeter())

shape2 = Shape(3, 4, 2)
print(shape2.area())
print(shape2.perimeter())

shape3 = Shape(3, 0, 2)
print(shape3.area())
print(shape3.perimeter())

shape4 = Shape(4, 7, -2)
print(shape4.area())
print(shape4.perimeter())

# Problem 7: Student Management Create a class representing a student with attributes like student ID, name, and marks. Implement methods to calculate the average grade and display student details.
class Student() :
    def __init__(self, id, name, *marks):
        self.id = id
        self.name = name
        self.marks = marks
    
    def display(self) :
        log.info("Displaying student information")
        return f"Name : {self.name}, ID : {self.id}, Marks : {self.marks}"
    
    def avgMarks(self) :
        log.info("Displaying avg marks")
        try : 
            for mark in self.marks :
                if mark < 0 :
                    raise Exception
            return sum(self.marks) / len(self.marks)
        except Exception :
            log.error("User entered invalid values")
            return "Enter Non negative numbers"

student1 = Student(5618, "Raghav", 90,85,87,70,100)
print(student1.display())
print(student1.avgMarks())
student2 = Student(5618, "Himanshu", 90,85,87,-70,100)
print(student2.display())
print(student2.avgMarks())
student3 = Student(5618, "Pratap", 90,85,87,"a",100)
print(student3.display())
print(student3.avgMarks())

# Problem 8: Email Management Create a class representing an email with attributes like sender, recipient, and subject. Implement methods to send an email and display email details.
class Email() :
    def __init__(self, sender, recipient, subject):
        self.sender = sender
        self.recipient = recipient
        self.subject = subject
    
    def display(self) :
        log.info("Displaying student information")
        return f"Recipient : {self.recipient}, Sender : {self.sender}, Subject : {self.subject}"
    
    def sendEmail(self) :
        a = len(self.recipient.split("@"))
        b = len(self.sender.split("@"))
        try :
            log.info("Checking if mail can be sent")
            if a == 1 or b == 1 or a > 2 or b > 2 :
                raise Exception
            if not self.recipient.split("@")[1] == "gmail.com" or not self.sender.split("@")[1] == "gmail.com" :
                raise Exception
        except Exception :
            log.error("Invalid Input")
            return "Invalid Input"
        else :
            log.info("Mail successfully sent")
            return "Mail successfully sent"
            
mail1 = Email("raghav@gmail.com", "abc@gmail.com", "Die")
print(mail1.display())
print(mail1.sendEmail())
mail2 = Email("raghavgmail.com", "abc@gmail.com", "Die")
print(mail2.display())
print(mail2.sendEmail())
mail3 = Email("raghav@gmail.com", "abc@gmail.com@123", "Die")
print(mail3.display())
print(mail3.sendEmail())

# Problem 9: Social Media Profile Create a class representing a social media profile with attributes like username and posts. Implement methods to add posts, display posts, and search for posts by keyword.
class SocialMediaProfile() :
    def __init__(self, username):
        self.username = username
        self.posts = []

    def add_post(self, post_content):
        try:
            if not isinstance(post_content, str):
                raise ValueError("Post content must be a string")
            self.posts.append(post_content)
            log.info(f"New post added by {self.username}")
        except ValueError as e:
            log.error(f"Error adding post by {self.username}: {e}")

    def display_posts(self):
        try:
            if self.posts:
                print(f"Posts by {self.username}:")
                for index, post in enumerate(self.posts, start=1):
                    print(f"{index}. {post}")
            else:
                print(f"No posts found for {self.username}")
        except Exception as e:
            log.error(f"Error displaying posts for {self.username}: {e}")

    def search_posts_by_keyword(self, keyword):
        try:
            matching_posts = [post for post in self.posts if keyword.lower() in post.lower()]
            if matching_posts:
                print(f"Posts by {self.username} containing '{keyword}':")
                for index, post in enumerate(matching_posts, start=1):
                    print(f"{index}. {post}")
            else:
                print(f"No posts found for {self.username} containing '{keyword}'")
        except Exception as e:
            log.error(f"Error searching posts for {self.username} with keyword '{keyword}': {e}")

profile1 = SocialMediaProfile("user123")
profile1.add_post("Hello world!")
profile1.add_post(123)  
profile1.add_post("Today is a great day.")
profile1.add_post("Python programming is fun.")
profile2 = SocialMediaProfile("user456")
profile2.add_post("Just finished reading a good book.")
profile2.add_post("Traveling to new places.")
profile1.display_posts()
print()
profile2.display_posts()
print()
profile1.search_posts_by_keyword("python")
print()
profile2.search_posts_by_keyword("travel")

# Problem 10: ToDo List Create a class representing a ToDo list with attributes like tasks and due dates. Implement methods to add tasks, mark tasks as completed, and display pending tasks.
import logging

class ToDoList:
    def __init__(self):
        self.tasks = {}

    def add_task(self, task_name, due_date):
        try:
            self.tasks[task_name] = due_date
            log.info(f"Added task '{task_name}' with due date '{due_date}'.")
        except Exception as e:
            log.error(f"Error adding task '{task_name}': {e}")

    def mark_completed(self, task_name):
        try:
            if task_name in self.tasks:
                del self.tasks[task_name]
                log.info(f"Marked task '{task_name}' as completed.")
            else:
                log.error(f"Task '{task_name}' not found in the ToDo list.")
        except Exception as e:
            log.error(f"Error marking task '{task_name}' as completed: {e}")

    def display_pending_tasks(self):
        try:
            if self.tasks:
                print("Pending Tasks:")
                for task, due_date in self.tasks.items():
                    print(f"{task} - Due Date: {due_date}")
            else:
                print("No pending tasks in the ToDo list.")
        except Exception as e:
            log.error(f"Error displaying pending tasks: {e}")
todo_list = ToDoList()
todo_list.add_task("Complete project proposal", "2024-03-15")
todo_list.add_task("Buy groceries", "2024-03-10")
todo_list.add_task("Finish coding assignment", "2024-03-20")
todo_list.display_pending_tasks()
todo_list.mark_completed("Buy groceries")
todo_list.display_pending_tasks()
```
